```
from google.colab import drive
drive.mount('/content/drive')


import sys
sys.path.append('/content/drive/MyDrive/DLS/project/packages')

import models

import torch

device = torch.device("cuda:0") if torch.cuda.is_available() else torch.device("cpu")
pool_images = None
pool_targets = None
gan = models.CycleGAN(lambd = 2, resnet_layers = 9, device = device, buffer_images=pool_images, dim_gen = 32, buffer_targets=pool_targets, dropout = 0.5)
gan.to(device)
checkpoint = torch.load('/content/drive/MyDrive/DLS/project/real2jojo_weights')
gan.load_state_dict(checkpoint['gan_state_dict'])
```

```
model = gan.gan_straight.generator
model.to('cpu')
torch.save({
      'state_dict': model.state_dict(),
      }, '/content/drive/MyDrive/DLS/project/generator_weights')
```
